```python
import json
import requests
import numpy as np

import time

import matplotlib.pyplot as plt
```

```python
imgs = []

for i in range(981):
    file = f"../images/test_{i}.JPEG"
    img = plt.imread(file).tolist()
    imgs.append(img)
```

```python
headers = {"content-type": "application/json"}
```

```python
t1 = time.time()
for i in range(981):
    print(i)
    json_response = requests.post('http://127.0.0.1:8501/v1/models/resnet152:predict', data=json.dumps({"signature_name": "serving_default", "instances": imgs[i:i+1]}), headers=headers)
print(time.time() - t1)
```
